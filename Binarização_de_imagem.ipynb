```python
from PIL import Image  # Biblioteca para tratamento da imagem
import matplotlib.pyplot as plt  # Biblioteca para visualização da imagem tratada
```

```python
imagem_caminho = '/content/lobo (2).jpg' # Caminho para imagem
```

```python
threshold_value = 127  # Limiar de binarização e pode se ajustar o quanto for necessário
```

```python
resultado_binario = binarizacao_imagem('/content/lobo (2).jpg', threshold_value)  # Binarização da imagem
```

```python
def binarizacao_imagem(imagem_caminho, threshold):
  image = Image.open(imagem_caminho).convert('L')  # Carregamento da imagem e conversão para escala de cinza
  pixels = list(Image.open(imagem_caminho).convert('L').getdata())  # Obter os pixels da imagem
  pixels_binarizados = [0 if pixel < threshold_value else 255 for pixel in pixels]  # Aplica a binarização
  imagem_binarizada = Image.new('L', Image.open(imagem_caminho).convert('L').size)  # Cria uma nova imagem binarizada
  imagem_binarizada.putdata(pixels_binarizados)

  return imagem_binarizada

```

```python
# Visualizar a imagem original e a imagem binarizarada

plt.subplot(121), plt.imshow(Image.open(imagem_caminho).convert('L'), 'gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(binarizacao_imagem('/content/lobo (2).jpg', threshold_value), 'gray')
plt.title('Imagem Binarizada'), plt.xticks([]), plt.yticks([])

plt.show()
```
